a. (1 point) Download the Child Mortality and the Income CSV files from https://www.gapminder.org/data/

b. (1 point) In your notebook, create an childmortality and income dataframes with the data from the corresponding CSV files.

```python
income = pd.read_csv("https://raw.githubusercontent.com/lanashin/Data_Viz/main/income_per_person_gdppercapita_ppp_inflation_adjusted.csv") 
```

```python
childmortality = pd.read_csv("https://raw.githubusercontent.com/lanashin/Data_Viz/main/child_mortality_0_5_year_olds_dying_per_1000_born.csv")
```

(3 points) For each of the three created dataframes, keep only the data that corresponds to 2020, and merge these three dataframes into one data dataframe that corresponds to the year 2020 and has four attributes: Country, ChildPerWoman, ChildMortality, Income. Merging dataframes corresponds to joining tables in SQL (https://pandas.pydata.org/docs/dev/user_guide/merging.html).

```python
#we have 3 dataframes: df, childmortality, income 
#we need just 2020 year 
df.drop(df.columns.difference(["country","2020"]), 1, inplace=True)
childmortality.drop(childmortality.columns.difference(["country","2020"]), 1, inplace=True)
income.drop(income.columns.difference(["country","2020"]), 1, inplace=True)
```

(2 points) Pay attention to the Income attribute, make sure that it can be used as an number (Hint: initially it cannot be used as a string because it contains the letter k for 1000).

```python
#replacing k with 00 and . with empty string
income["2020"] = income ["2020"].str.replace("k","00")
income["2020"] = income["2020"].str.replace(".","").astype(float)
```

```python
income.head()
```

```python
# all dataframes have four attributes: Country, ChildPerWoman, ChildMortality, Income
df.rename(columns= {"2020" : "ChildPerWoman"},inplace = True)
income.rename(columns= {"2020" : "Income"}, inplace = True)
childmortality.rename(columns= {"2020" : "ChildMortality"}, inplace = True)
```

```python
#Merging all 3 dataframes
frames = [df, income , childmortality]

#result = pd.concat(frames)
result = pd.concat([df, income , childmortality], axis=1, join="inner")
result
```

```python
#to change index that will fit all dataframes when I join them together
merging = pd.DataFrame()
df.set_index("country", inplace = True)
income.set_index("country", inplace =True)
childmortality.set_index("country", inplace = True)
result = df.join([income, childmortality], how = 'inner')
result
```
